Deteccion de valores atípicos

```python
import matplotlib.pyplot as plt
plt.title('Edad')
plt.hist(data_clean['Age'], edgecolor = 'black', linewidth=1)
plt.show() 
```

```python
plt.title('SibSp')
plt.hist(data_clean['SibSp'], edgecolor = 'black', linewidth=1)
plt.show() 
```

```python
plt.boxplot(data_clean['SibSp'],vert=False)
plt.show()
```

```python
Q1 = data_clean['SibSp'].quantile(0.25)
print("Primer Cuartil",Q1)
Q3 = data_clean['SibSp'].quantile(0.75)
print("Tercer Cuartil",Q3)

IQR = Q3 - Q1
print("Rango Intercuartil",IQR)
Mediana = data_clean['SibSp'].median()
print("Mediana",Mediana)

Valor_Min = data_clean['SibSp'].min()
print("Valor minimo",Valor_Min)

Valor_Max = data_clean['SibSp'].max()
print("Valor minimo",Valor_Max)
```

```python
BI_Calculado = (Q1-1.5 * IQR)
print("BI_Calculado\n", BI_Calculado)

BS_Calculado = (Q3+1.5 * IQR)
print("BS_Calculado\n", BS_Calculado)
```

Hallamos los valores outliers 

```python
outliers = ((data_clean['SibSp']<BI_Calculado) | (data_clean['SibSp']>BS_Calculado ))
valor = data_clean[outliers]
print(valor["SibSp"])
```

Eliminar los valores outliers 

```python
ubicacion_sin_out = (data_clean['SibSp'] >= BI_Calculado) & (data_clean['SibSp'] <= BS_Calculado)
data_sin_out = data_clean[ubicacion_sin_out]
data_sin_out
```

```python
plt.boxplot(data_sin_out['SibSp'],vert=False)
plt.show()
```

Cambiar variables categóricas por numéricas 

```python
data_sin_out.info()
```

```python
data_sin_out['Sex'] = pd.factorize(data_sin_out['Sex'])[0]
data_sin_out
```

Agrupamiento de datos 

Agruparemos las edades por 
   0 - 5 , 6 - 12 , 13 - 18 , 19 - 25 , 36 - 60 , 60 - 100 

```python
bins = [0,5,12,18,35,60,100]
names = ["1","2","3","4","5","6"]
data_sin_out["Age"] = pd.cut(data_sin_out["Age"],bins,labels=names)
```
